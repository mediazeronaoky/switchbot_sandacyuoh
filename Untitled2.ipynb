```python
import requests
import pandas as pd
import datetime
import time

# SwitchBot APIのエンドポイントとAPIキー
API_URL = "https://api.switch-bot.com/v1.0/devices/{device_id}/status"
API_TOKEN = "f99528c5b951b68df7a4316bbc33eadc0573ecf3a8a7505323b499db96c0be9c90e05dfa5b1782b5c6544eef082badc6"  # ここにAPIトークンを入力
DEVICE_ID = "CE1706066C27"  # ここにデバイスIDを入力

# ヘッダー情報
headers = {
    "Authorization": API_TOKEN,
    "Content-Type": "application/json; charset=utf8",
}

# データ取得と保存の関数
def fetch_and_save_data():
    response = requests.get(API_URL.format(device_id=DEVICE_ID), headers=headers)

    if response.status_code == 200:
        data = response.json()
        temperature = data['body']['temperature']
        humidity = data['body']['humidity']
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # DataFrameに変換
        df = pd.DataFrame({
            "Timestamp": [timestamp],
            "Temperature": [temperature],
            "Humidity": [humidity]
        })

        # データをCSVファイルに保存
        df.to_csv('temperature_humidity_data.csv', mode='a', header=not pd.io.common.file_exists('temperature_humidity_data.csv'), index=False)

    else:
        print(f"Failed to retrieve data: {response.status_code}")

# データを一定間隔で取得（例: 48回、30分間隔）
for i in range(48):
    fetch_and_save_data()
    time.sleep(1800)  # 1800秒（30分）待機
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルからデータを読み込む
df = pd.read_csv('temperature_humidity_data.csv')

# タイムスタンプを日時形式に変換
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# グラフの作成
plt.figure(figsize=(10, 5))

# 温度のプロット
plt.plot(df['Timestamp'], df['Temperature'], label='Temperature (°C)', color='red')

# 湿度のプロット
plt.plot(df['Timestamp'], df['Humidity'], label='Humidity (%)', color='blue')

# グラフの装飾
plt.title('Temperature and Humidity over Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# グラフの表示
plt.tight_layout()
plt.show()
```
